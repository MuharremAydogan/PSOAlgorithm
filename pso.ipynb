```python
import math
import random
```

### Let's minimize the x^2+y^2 algorithm with the PSO algorithm

```python
def amac_fonksiyonu(konum):
    x, y = konum
    return x**2 + y**2  
```

```python
class Parcacik:
    def __init__(self):
        self.konum=[random.uniform(-10, 10), random.uniform(-10, 10)]
        self.hiz=[random.uniform(-1, 1), random.uniform(-1, 1)]
        self.en_iyi_konum=list(self.konum)
        self.en_iyi_skor=amac_fonksiyonu(self.konum)
    def hizi_guncelle(self, kumes_en_iyi_konum, w=0.5, c1=1.5, c2=1.5):
        
        for i in range(2):
            r1=random.random()  # bireysel etki için rastgelelik
            r2 = random.random()  # sosyal etki için rastgelelik
            bireysel=c1*r1*(self.en_iyi_konum[i] - self.konum[i])
            sosyal =c2*r2*(kumes_en_iyi_konum[i] - self.konum[i])
            
            self.hiz[i] = w * self.hiz[i] + bireysel + sosyal    #yeni hız= eskihız*0.5+bireysel+sosyal

    
    def konumu_guncelle(self):
        for i in range(2):
            self.konum[i]+=self.hiz[i]  #yeni konum = eski konum + hız     (normalde hız*zaman fakat burada zaman ogesi olmadıgı icin zaman=1 gibi düşünebiliriz)

        mevcut_skor=amac_fonksiyonu(self.konum)
        if mevcut_skor<self.en_iyi_skor:
            self.en_iyi_skor=mevcut_skor
            self.en_iyi_konum=list(self.konum)
```

```python
def pso(parcacik_sayisi=30, iterasyon_sayisi=100):
    suru=[Parcacik() for _ in range(parcacik_sayisi)]

    
    kumes_en_iyi_konum=min(suru, key=lambda p: p.en_iyi_skor).en_iyi_konum
    kumes_en_iyi_skor=amac_fonksiyonu(kumes_en_iyi_konum)

    for iterasyon in range(iterasyon_sayisi):
        for parcacik in suru:
            parcacik.hizi_guncelle(kumes_en_iyi_konum)
            parcacik.konumu_guncelle()

        mevcut_en_iyi=min(suru, key=lambda p: p.en_iyi_skor)
        if mevcut_en_iyi.en_iyi_skor<kumes_en_iyi_skor:
            kumes_en_iyi_skor=mevcut_en_iyi.en_iyi_skor
            kumes_en_iyi_konum=list(mevcut_en_iyi.en_iyi_konum)

        print(f"iterasyon {iterasyon + 1}: en iyi skor={kumes_en_iyi_skor:.5f}")

    
    print("\nEn iyi bulunan çözüm:")
    print("konum:", kumes_en_iyi_konum)
    print("skor(min hata):", kumes_en_iyi_skor)
```

```python
pso()
```
